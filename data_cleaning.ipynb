```python
def convert_to_lakhs(price):
    price = price.replace('₹', '').replace(',', '').strip()
    if 'Crore' in price:
        return float(price.replace('Crore', '').strip()) * 100
    elif 'Lakh' in price:
        return float(price.replace('Lakh', '').strip())     
    else:
        return float(price)/100
```

```python
df.loc[:,'price_in_lakhs'] = df['price'].apply(convert_to_lakhs)
```

```python
import re 
def clean_wheel_size(size):
  if pd.isna(size):  # Check for missing values (NaN)
    return size
  pattern = r"^R?(\d+)"  # Match optional 'R' followed by one or more digits
  match = re.search(pattern, size)
  if match:
    return match.group(1)  # Extract the captured digit group
  else:
    return size  # Return the original value if no match is found

df['Wheel Size'] = df['Wheel Size'].apply(clean_wheel_size)
```

```python
def clean_top_speed(top_speed):
    pattern = r'\d{3,5}'
    match = re.search(pattern, top_speed)
    if match and match.group(0):
        return match.group(0)
    else:
        return top_speed
df['Top Speed'] = df['Top Speed'].astype(str).apply(clean_top_speed)
```

```python
def extract_numbers(value):
    num = re.findall(r'\d+\.?\d*', str(value))  # Find all numbers (including decimals)
    return float(num[0]) if num else None  # Convert to float
df['Turning Radius'] = df['Turning Radius'].apply(extract_numbers)
```

```python
def clean_super_charger(x):
    if pd.isna(x):  # Check if x is NaN
        return np.nan
    elif x in ['No', 'NO', 'no']:
        return 'No'
    elif x == 'yes':
        return 'Yes'
    else:
        return 'Yes'
    
df['Super Charger'] = df['Super Charger'].apply(clean_super_charger)
```

```python

def extract_numbers_cargo(value):
    num = re.findall(r'\d+\.?\d*', str(value))  # Extract all numeric values
    if num:  
        return float(num[0])  # Convert the first found number to float
    return None  # Return None if no numbers are found

df['Cargo Volumn'] = df['Cargo Volumn'].apply(extract_numbers_cargo)
```

```python
def clean_turbo_charger(x):
    if pd.isna(x):  # Check if x is NaN
        return np.nan
    elif x in ['No', 'NO', 'no']:
        return 'No'
    elif x in ['Yes', 'YES', 'yes']:
        return 'Yes'
    elif x in ['Twin', 'twin']:
        return 'Yes'
    else:
        return x
    
df['Turbo Charger'] = df['Turbo Charger'].apply(clean_turbo_charger)
```

```python
import numpy as np

def clean_drive_type(value):
    if pd.isna(value):  
        return np.nan  # Keep NaN values
    
    value = str(value).strip().lower()  # Normalize case and remove spaces
    
    # Mapping similar values to standardized categories
    if value in ['fwd', 'front wheel drive']:
        return 'FWD'
    elif value in ['rwd', 'rear wheel drive with esp', 'rwd(with mtt)']:
        return 'RWD'
    elif value in ['awd', 'all wheel drive', 'permanent all-wheel drive quattro', 'awd integrated management']:
        return 'AWD'
    elif value in ['4wd', '4 wd', '4x4', '4x4']:
        return '4WD'
    elif value in ['2wd', '2 wd', '4x2', '4x2', 'two wheel drive']:
        return '2WD'
    else:
        return value  # Return unchanged if not found in mapping

# Apply the function to the DataFrame
df['Drive Type'] = df['Drive Type'].apply(clean_drive_type)
```

```python
def clean_insurance(x):
    if x == 'Third Party':
        return 'Third Party insurance'
    elif pd.isna(x):
        return 'Unknown'
    elif x in['2','1']:
        return 'Unknown'
    else:
        return x

df['Insurance Validity'] = df['Insurance Validity'].apply(clean_insurance)
```

```python
def standardize_tyre_type(value):
    corrections = {
        "tubeless tyre": "tubeless",
        "tubeless tyres": "tubeless",
        "tubless": "tubeless",
        "tubless, radial": "tubeless, radial",
        "radial, tubless": "tubeless, radial",
        "tubeless,radials": "tubeless, radial",
        "tubeless radial tyres": "tubeless, radial",
        "radial tubeless": "tubeless, radial",
        "tubeless tyres all terrain": "tubeless, all terrain",
        "run-flat": "runflat",
        "runflat tyres": "runflat",
        "runflat tyre": "runflat",
        "runflat,radial": "runflat, radial",
        "radial tyres": "radial"
    }

    if pd.isna(value) or str(value).lower() in ["nan", "none"]:
        return "unknown"
    
    value = str(value).lower().strip()
    value = value.replace(", ", ",")  # Remove extra spaces after commas
    
    return corrections.get(value, value)

# Apply it correctly to a single column
df["Tyre Type"] = df["Tyre Type"].apply(standardize_tyre_type)
```
